```python
# -*- coding: UTF-8 -*-
"""PyBank Homework."""

# Import libraries:
from pathlib import Path
import csv

# Set file path for budget_data.csv:
budget_filepath = Path('budget_data.csv')

# Initialize variables:
average_pnl_change = 0
current_pnl_change = 0
financial_analysis = []
last_pnl_line = 0
max_pnl_increase_amount = float('-inf')
"""
NOTE: In class, we were taught to initialize numeric variables with a '0';
This could be problematic, depending on the dataset to be analyzed and how it is to be analyzed.
I opted to initialize the above variable with a negative infinity float so that any 'pnl increase' could be considered larger than the initialized value.
"""
max_pnl_increase_date = ""
max_pnl_decrease_amount = float('inf')
"""
NOTE: In class, we were taught to initialize numeric variables with a '0';
This could be problematic, depending on the dataset to be analyzed and how it is to be analyzed.
I opted to initialize the above variable with a positive infinity float so that any 'pnl decrease' could be considered smaller than the initialized value.
"""
max_pnl_decrease_date = ""
total_months = 0
total_net_pnl = 0

# Open budget_data.csv:
with open(budget_filepath, newline = '') as file:
    budget_data = csv.reader(file, delimiter=',')
    
    # Ignore header row:
    next(budget_data)
    
    """
    NOTE: Instead of writing additional (some duplicated) code to work with the first record of our data outside of the following 'for' loop (which contains calculations for all proceeding records), 
    I've opted to utilize the 'enumerate' function so I can assign specific code to all records except for the first.
    We were not taught the 'enumerate' function in class, though I've learned of it in my studies.
    """
    
    # Iterate over dataset, enumerating rows:
    for row in enumerate(budget_data, start = 0):
        
        # Tally months in dataset:
        total_months += 1
        
        # Tally net total pnl over period:
        total_net_pnl += int(row[1][1])
        
        # Prepare to calculate differences between current row and previous row:
        if row[0] != 0:
            
            # Calculate and add current pnl change to average_pnl_change:
            current_pnl_change = int(row[1][1]) - last_pnl_line
            average_pnl_change += current_pnl_change
            
            # Calculate current change in profit; set to max pnl decrease or max pnl increase, if appropriate:
            if current_pnl_change < max_pnl_decrease_amount:
                max_pnl_decrease_amount = current_pnl_change
                max_pnl_decrease_date = row[1][0]
            elif current_pnl_change > max_pnl_increase_amount:
                max_pnl_increase_amount = current_pnl_change
                max_pnl_increase_date = row[1][0]
        
        # Set current pnl to last_pnl_line:
        last_pnl_line = int(row[1][1])

# Calculate average pnl change over period:
average_pnl_change = average_pnl_change / (total_months - 1)

# Add financial analysis lines to list:
financial_analysis = ["Financial Analysis", "----------------------------", "Total Months: " + str(total_months), "Total: $" + str(total_net_pnl), f"Average Change: ${round(average_pnl_change, 2)}", "Greatest Increase in Profits: " + max_pnl_increase_date + " ($" + str(max_pnl_increase_amount) + ")", "Greatest Decrease in Profits: " + max_pnl_decrease_date + " ($" + str(max_pnl_decrease_amount) + ")"]

# Print financial analysis to terminal:
for line in financial_analysis:
    print (line)

# Write financial analysis to README.txt file:
with open('README.txt', 'w') as f:
    for line in financial_analysis:
        f.write(line)
        f.write("\n")

"""
The following is not a part of the challenge.
I've opted to create a markdown file for display in my Github repository.
"""

# Write financial analysis to README.md file:
with open('README.md', 'w') as f:
    for line in financial_analysis:
        f.write(line + "  ")
        f.write("\n")
```
